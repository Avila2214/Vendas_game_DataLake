```python
import pandas as pd
```

```python
# salvando o arquivo csv em um df
df = pd.read_csv('../bronze/Video_Games_Sales_as_at_22_Dec_2016.csv')

df.head(5)
```

```python
# Verificando valores nulos e fazendo a contagem.

df.isnull().sum()
```

```python
# Alterando os nomes das colunas

df = df.rename(columns={
    'Name' : 'Nome_Jogo',
    'Platform' : 'Plataforma',
    'Year_of_Release' : 'Ano_de_Lancamento',
    'Genre' : 'Genero',
    'Publisher' : 'Editora',
    'NA_Sales' : 'NA_Vendas',
    'EU_Sales' : 'EU_Vendas',
    'JP_Sales' : 'JP_Vendas',
    'Other_Sales' : 'Outras_Vendas',
    'Global_Sales' : 'Vendas_Globais',
    'Critic_Score' : 'Pontuacao_Critica',
    'Critic_Count' : 'Contagem_Critica',
    'User_Score' : 'Pontuacao_do_usuario',
    'User_Count' : 'Contagem_de_usuarios',
    'Developer' : 'Desenvolvedor',
    'Rating' : 'Classificacao'
})

df
```

```python
# Alterando os valores da coluna 'Genero'

df['Genero'] = df['Genero'].replace({
    'Sports': 'Esporte',
    'Platform': 'Plataforma',
    'Racing': 'Corrida',
    'Role-Playing': 'RPG',
    'Puzzle': 'Quebra_Cabeca',
    'Misc': 'Diversos',
    'Shooter': 'Tiro',
    'Simulation': 'Simulação',
    'Action': 'Acao',
    'Fighting': 'Combate',
    'Adventure': 'Aventura',
    'Strategy': 'Estratégia'
})

df
```

```python
# Alterando os valores nulos para 'Nao_inserido' 

df = df.fillna('Nao_Inserido')
df
```

```python
# Verificando valores nulos

df.isnull().sum()
```

```python
# Verificando os valores da coluna 'Classificacao'

df['Classificacao'].unique()
```

```python
# Alerando os valores da coluna 'Classificacao'

df['Classificacao'] = df['Classificacao'].replace({
    'E': 'Livre',
    'E10+': '+10',
    'T': '+14',
    'M': '+17',
    'AO': '+18',
    'EC': 'Infantil',
    'K-A': 'Publico Geral',
    'RP': 'Aguardando Classificação'
})

df
```

```python
# Colunas numéricas para float

colunas_float = ['NA_Vendas', 'EU_Vendas', 'JP_Vendas', 
                 'Outras_Vendas', 'Vendas_Globais', 'Pontuacao_Critica', 
                 'Pontuacao_do_usuario']

df[colunas_float] = df[colunas_float].apply(pd.to_numeric, errors='coerce')

df.info()
```

```python
# Colunas numericas para int

df['Ano_de_Lancamento'] = pd.to_numeric(df['Ano_de_Lancamento'], errors= 'coerce').fillna(0).astype(int)
df['Contagem_Critica'] = pd.to_numeric(df['Contagem_Critica'], errors='coerce').fillna(0).astype(int)
df['Contagem_de_usuarios'] = pd.to_numeric(df['Contagem_de_usuarios'], errors='coerce').fillna(0).astype(int)
```

```python
# Remover espaços extras nos nomes

df['Nome_Jogo'] = df['Nome_Jogo'].str.strip()
df['Genero'] = df['Genero'].str.capitalize()
df['Editora'] = df['Editora'].str.strip()
df['Desenvolvedor'] = df['Desenvolvedor'].str.strip()
```

```python
# Removendo duplicatas

df.drop_duplicates(inplace=True)
```

```python
# Removendo linhas em que todas as vendas estão zeradas

df = df[~((df[['NA_Vendas', 'EU_Vendas', 'JP_Vendas', 'Outras_Vendas']] == 0).all(axis=1))]
```

```python
# Remover anos fora da faixa válida (1970 ate 2030): 

df = df[(df['Ano_de_Lancamento'] >= 1970) & (df['Ano_de_Lancamento'] <= 2030)]

df
```

```python
# Exibir resumo estatístico e possíveis valores ausentes

print(df.info())
print(df.isna().sum())
```

```python
df.to_csv('Limpo_Video_Games_Sales_as_at_22_Dec_2016.csv')
```

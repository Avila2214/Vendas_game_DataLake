```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../silver/Silver_Video_Games_Sales_as_at_22_Dec_2016.csv')
```

```python

# Garantir que as colunas de pontuações sejam do tipo numérico
numeric_columns = ["Pontuacao_Critica", "Pontuacao_do_usuario"]
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Preencher pontuações ausentes com a média das existentes
df["Pontuacao_Critica"].fillna(df["Pontuacao_Critica"].mean(), inplace=True)
df["Pontuacao_do_usuario"].fillna(df["Pontuacao_do_usuario"].mean(), inplace=True)
```

```python
# 2. Padronização de categorias
df["Genero"] = df["Genero"].str.capitalize()
```

```python
# Faixa Etária
df["Faixa_Etaria"] = np.where(df["Classificacao"] == "Livre", "Todos", "Restrito")

# Sucesso Comercial
df["Sucesso_Comercial"] = np.select(
    [
        df["Vendas_Globais"] > 50,
        df["Vendas_Globais"] > 20
    ],
    ["Alto", "Médio"],
    default="Baixo"
)
```

```python
# Cálculo de percentuais regionais
df["Percentual_NA"] = (df["NA_Vendas"] / df["Vendas_Globais"]) * 100
df["Percentual_EU"] = (df["EU_Vendas"] / df["Vendas_Globais"]) * 100
df["Percentual_JP"] = (df["JP_Vendas"] / df["Vendas_Globais"]) * 100
df["Percentual_Outras"] = (df["Outras_Vendas"] / df["Vendas_Globais"])
```

```python
# 5. Ranking por Gênero
df["Ranking_Genero"] = df.groupby("Genero")["Vendas_Globais"].rank(ascending=False, method="min")
```

```python
# Exibir DataFrame final
print("Dados refinados para camada Gold:")
print(df.head())
```
